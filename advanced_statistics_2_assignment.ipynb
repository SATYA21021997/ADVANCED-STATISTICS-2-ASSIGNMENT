1.  What is hypothesis testing in statistics
?

ANS - Hypothesis testing is a fundamental method used in statistics to make inferences or decisions about a population based on sample data.

2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?


ANS - The null hypothesis (H₀) and the alternative hypothesis (H₁ or Ha) are the two competing statements in hypothesis testing.

Null Hypothesis (H₀)
The null hypothesis is a statement of no effect, no difference, or the status quo. It serves as the default assumption that there is no relationship between variables or that any observed effect is due to random chance.

Example: If we are testing whether a new drug is effective, the null hypothesis might be "The new drug has no effect on the disease."

Alternative Hypothesis (H₁ or Ha)
The alternative hypothesis is a statement that suggests a potential effect, difference, or relationship that we aim to detect or prove. It represents the claim we are testing for and is considered the opposite of the null hypothesis.

Example: In the same drug effectiveness test, the alternative hypothesis might be "The new drug is effective in treating the disease."

3.  What is the significance level in hypothesis testing, and why is it important?


ANS - The significance level (α) is a crucial concept in hypothesis testing. It represents the threshold probability at which we are willing to reject the null hypothesis.

4.  What does a P-value represent in hypothesis testing
?

ANS- The p-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis is true. In other words, it measures how well the sample data supports the null hypothesis.

A smaller p-value indicates stronger evidence against the null hypothesis.

5.  How do you interpret the P-value in hypothesis testing
?

ANS - Steps for Interpretation
Compare the P-value with the Significance Level (α):

The first step in interpreting the p-value is to compare it with the significance level (α), which is often set at 0.05, 0.01, or 0.10.

Decision Rule:

If p-value ≤ α: Reject the null hypothesis (H₀).

If p-value > α: Fail to reject the null hypothesis (H₀).

6.  What are Type 1 and Type 2 errors in hypothesis testing
?

ANS - Type I Error (False Positive)
Definition: A Type I error occurs when the null hypothesis (H₀) is rejected when it is actually true. In other words, we conclude that there is an effect or difference when there really isn't one.

Significance Level (α): The probability of making a Type I error is denoted by the significance level (α). Commonly used values for α are 0.05, 0.01, or 0.10.

Example: Suppose we test a new drug's effectiveness and conclude that it works, when in reality, it doesn't. This is a Type I error.

Type II Error (False Negative)
Definition: A Type II error occurs when the null hypothesis (H₀) is not rejected when it is actually false. In other words, we fail to detect an effect or difference when there actually is one.

Power of the Test (1 - β): The probability of making a Type II error is denoted by β. The power of the test, which is the probability of correctly rejecting the null hypothesis, is given by (1 - β).

Example: Suppose we test a new drug's effectiveness and conclude that it doesn't work, when in reality, it does. This is a Type II error.

7.  What is the difference between a one-tailed and a two-tailed test in hypothesis testing
?

ANS - The primary difference between a one-tailed and a two-tailed test in hypothesis testing lies in the direction of the test and what we are trying to detect regarding the effect or difference.

One-Tailed Test
Definition: A one-tailed test (also known as a directional test) examines if a parameter is either greater than or less than a certain value, but not both.

Hypotheses: The alternative hypothesis (H₁) specifies a direction of the effect (e.g., greater than or less than).

Example: Suppose we are testing if a new drug is more effective than an existing one.

Null Hypothesis (H₀): The new drug is not more effective than the existing one.

Alternative Hypothesis (H₁): The new drug is more effective than the existing one.

Use Case: One-tailed tests are used when we have a specific direction in mind and are only interested in deviations in that direction.

Critical Region: The entire significance level (α) is concentrated in one tail of the distribution (either left or right).

Two-Tailed Test
Definition: A two-tailed test (also known as a non-directional test) examines if a parameter is different from a certain value, without specifying a direction.

Hypotheses: The alternative hypothesis (H₁) does not specify a direction but indicates a difference in either direction (e.g., not equal to).

Example: Suppose we are testing if a new drug is different from an existing one.

Null Hypothesis (H₀): The new drug is not different from the existing one.

Alternative Hypothesis (H₁): The new drug is different from the existing one.

10.  What is the T-distribution, and when should it be used instead of the normal distribution
?

ANS - The t-distribution, also known as Student's t-distribution, is a probability distribution that is used in hypothesis testing and in constructing confidence intervals, particularly when dealing with small sample sizes or unknown population variances. Here's a closer look:

What is the T-distribution?
Shape: The t-distribution is similar to the normal distribution but has heavier tails, meaning it has a higher probability for extreme values.

Degrees of Freedom (df): The shape of the t-distribution is determined by the degrees of freedom, which is related to the sample size. As the sample size increases, the t-distribution approaches the normal distribution.

When to Use the T-distribution
Small Sample Sizes: The t-distribution is particularly useful when the sample size is small (typically less than 30).

Unknown Population Variance: It is used when the population variance is unknown and needs to be estimated from the sample data.

Normality Assumption: The t-distribution assumes that the data follows a normal distribution or approximately normal for smaller samples.

Comparison with the Normal Distribution
Sample Size:

T-distribution: Used for small sample sizes (n < 30).

Normal distribution (Z-distribution): Used for large sample sizes (n ≥ 30).

Population Variance:

T-distribution: Population variance is unknown.

Normal distribution: Population variance is known.

Degrees of Freedom:

T-distribution: Varies with degrees of freedom (df = n - 1, where n is the sample size).

Normal distribution: Fixed shape (no degrees of freedom parameter).

The variances within each group should be approximately equal. This can be tested using Levene's test or Bartlett's test.

21.  What are the different types of ANOVA tests
?

ANS - ANOVA, or Analysis of Variance, encompasses several types, each suited to specific experimental designs and research questions. Here are the main types of ANOVA tests:

1. One-Way ANOVA
Purpose: Used to compare the means of three or more groups based on a single independent variable.

Example: Comparing the average test scores of students taught using three different teaching methods.

2. Two-Way ANOVA
Purpose: Used to compare the means of groups based on two independent variables and their interaction.

Example: Assessing the effects of different teaching methods and study durations on student test scores.

3. Repeated Measures ANOVA
Purpose: Used when the same subjects are measured multiple times under different conditions.

Example: Measuring the effects of different diets on the same group of individuals over time.

4. Mixed-Design ANOVA (Split-Plot ANOVA)
Purpose: Combines between-subjects and within-subjects factors in the analysis.

Example: Studying the effects of different teaching methods (between-subjects factor) on students measured at multiple time points (within-subjects factor).

5. Multivariate Analysis of Variance (MANOVA)
Purpose: Extends ANOVA to handle multiple dependent variables simultaneously.

Example: Comparing the effects of different therapies on multiple health outcomes (e.g., blood pressure, cholesterol levels).

6. Analysis of Covariance (ANCOVA)

22.  What is the F-test, and how does it relate to hypothesis testing?


ANS - The F-test is a statistical test used to compare variances and assess the overall significance in multiple regression models or analysis of variance (ANOVA). It is particularly useful when comparing the means of three or more groups to determine if there are significant differences among them.

What is the F-test?
Definition: The F-test evaluates the ratio of two variances to determine if they are significantly different. It follows the F-distribution, which is positively skewed and depends on the degrees of freedom of the numerator and denominator.

Formula: The F-statistic is calculated as: $$ F = \frac{\text{Variance between groups (MSB)}}{\text{Variance within groups (MSW)}} $$

How the F-test Relates to Hypothesis Testing
Formulate Hypotheses:

Null Hypothesis (H₀): The group variances are equal (no significant difference between group means).

Alternative Hypothesis (H₁): At least one group variance is different (significant difference between group means).

Calculate the F-statistic:

1.  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    std_error = population_std / np.sqrt(sample_size)

    z_score = (sample_mean - population_mean) / std_error

    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    reject_null = p_value < alpha

    return z_score, p_value, reject_null

sample_data = [10, 12, 9, 11, 13, 10, 12, 8, 14, 10]
population_mean = 10
population_std = 2

z_score, p_value, reject_null = z_test(sample_data, population_mean, population_std)

print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")
print("Reject Null Hypothesis" if reject_null else "Fail to Reject Null Hypothesis")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_and_visualize_errors(sample_mean, population_mean, population_std, sample_size, alpha=0.05, alternative_mean=None):
    z_critical = stats.norm.ppf(1 - alpha / 2)

    std_error = population_std / np.sqrt(sample_size)

    z_score = (sample_mean - population_mean) / std_error

    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    reject_null = p_value < alpha

    x = np.linspace(population_mean - 4*population_std, population_mean + 4*population_std, 1000)
    null_dist = stats.norm.pdf(x, population_mean, std_error)

    if alternative_mean is None:
        alternative_mean = sample_mean
    alternative_dist = stats.norm.pdf(x, alternative_mean, std_error)

    plt.plot(x, null_dist, label='Null Hypothesis Distribution')
    plt.plot(x, alternative_dist, label='Alternative Hypothesis Distribution')

    plt.fill_between(x, 0, null_dist, where=(x >= population_mean + z_critical * std_error) | (x <= population_mean - z_critical * std_error),
                     color='gray', alpha=0.5, label='Type 1 Error Region')

    plt.fill_between(x, 0, alternative_dist, where=(x < population_mean + z_critical * std_error) & (x > population_mean - z_critical * std_error),
                     color='lightblue', alpha=0.5, label='Type 2 Error Region')

    plt.axvline(sample_mean, color='red', linestyle='--', label=f'Sample Mean: {sample_mean}')

    plt.title('Type 1 and Type 2 Errors')
    plt.xlabel('Mean Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

    return z_score, p_value, reject_null

sample_mean = 11
population_mean = 10
population_std = 2
sample_size = 30
alpha = 0.05

z_score, p_value, reject_null = calculate_and_visualize_errors(sample_mean, population_mean, population_std, sample_size, alpha)

print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")
print("Reject Null Hypothesis" if reject_null else "Fail to Reject Null Hypothesis")
```

9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.


```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    std_error = sample_std / np.sqrt(sample_size)

    if sample_size > 30:
        z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
        margin_of_error = z_critical * std_error
    else:
        t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, df=sample_size-1)
        margin_of_error = t_critical * std_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

sample_data = [23, 20, 25, 22, 21, 24, 20, 19, 22, 21]
confidence_level = 0.95

lower_bound, upper_bound = confidence_interval(sample_data, confidence_level)

print(f"Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

def bayesian_update(prior_alpha, prior_beta, data):
    heads = np.sum(data)
    tails = len(data) - heads

    posterior_alpha = prior_alpha + heads
    posterior_beta = prior_beta + tails

    return posterior_alpha, posterior_beta

prior_alpha = 1
prior_beta = 1

data = [1, 0, 1, 1, 0, 1, 1, 0, 1, 1]

posterior_alpha, posterior_beta = bayesian_update(prior_alpha, prior_beta, data)

x = np.linspace(0, 1, 1000)
prior_dist = beta(prior_alpha, prior_beta).pdf(x)
posterior_dist = beta(posterior_alpha, posterior_beta).pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, prior_dist, label=f'Prior: Beta({prior_alpha}, {prior_beta})', linestyle='--')
plt.plot(x, posterior_dist, label=f'Posterior: Beta({posterior_alpha}, {posterior_beta})')
plt.xlabel('Probability of Heads')
plt.ylabel('Density')
plt.title('Bayesian Update of Coin Bias')
plt.legend()
plt.show()
```

16.  Implement an F-test using Python to compare the variances of two random samples


```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    f_statistic = var1 / var2

    df1 = len(sample1) - 1
    df2 = len(sample2) - 1


    p_value = stats.f.cdf(f_statistic, df1, df2)
    p_value = 2 * min(p_value, 1 - p_value)  # Two-tailed test

    return f_statistic, p_value


np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=2, size=30)
sample2 = np.random.normal(loc=10, scale=3, size=30)

f_statistic, p_value = f_test(sample1, sample2)

print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")
print("Reject Null Hypothesis" if p_value < 0.05 else "Fail to Reject Null Hypothesis")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def plot_f_distribution(df1, df2):
    x = np.linspace(0, 5, 1000)
    y = f.pdf(x, df1, df2)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})')
    plt.fill_between(x, 0, y, alpha=0.2)
    plt.title('F-distribution')
    plt.xlabel('F-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()


df1 = 5
df2 = 10
plot_f_distribution(df1, df2)
```

22.  Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def anova_test(*groups):

    f_statistic, p_value = stats.f_oneway(*groups)
    return f_statistic, p_value


group1 = [23, 20, 25, 22, 21, 24, 20, 19, 22, 21]
group2 = [30, 28, 32, 29, 31, 30, 28, 27, 29, 30]
group3 = [35, 34, 36, 33, 34, 35, 36, 32, 33, 34]

f_statistic, p_value = anova_test(group1, group2, group3)

print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")
print("Reject Null Hypothesis" if p_value < 0.05 else "Fail to Reject Null Hypothesis")


all_data = [group1, group2, group3]
labels = ['Group 1', 'Group 2', 'Group 3']


plt.figure(figsize=(10, 6))
plt.boxplot(all_data, labels=labels)
plt.title('Box Plots of Different Groups')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()

group_means = [np.mean(group) for group in all_data]

plt.figure(figsize=(10, 6))
plt.bar(labels, group_means, color=['blue', 'green', 'red'])
plt.title('Mean Values of Different Groups')
plt.xlabel('Groups')
plt.ylabel('Mean Values')
plt.show()
```

23.  Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)
mean1 = 50
std1 = 10
n1 = 30
mean2 = 55
std2 = 12
n2 = 30

sample1 = np.random.normal(mean1, std1, n1)
sample2 = np.random.normal(mean2, std2, n2)


t_statistic, p_value = stats.ttest_ind(sample1, sample2)


print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")
print("Reject Null Hypothesis" if p_value < 0.05 else "Fail to Reject Null Hypothesis")


plt.figure(figsize=(12, 6))

plt.hist(sample1, bins=15, alpha=0.5, label='Sample 1', color='blue')
plt.hist(sample2, bins=15, alpha=0.5, label='Sample 2', color='orange')

plt.axvline(np.mean(sample1), color='blue', linestyle='dashed', linewidth=1)
plt.axvline(np.mean(sample2), color='orange', linestyle='dashed', linewidth=1)

plt.title('Histogram of Simulated Data from Normal Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
import scipy.stats as stats

def chi_square_variance_test(sample_data, hypothesized_variance, alpha=0.05):
    sample_size = len(sample_data)
    sample_variance = np.var(sample_data, ddof=1)


    chi_square_statistic = (sample_size - 1) * sample_variance / hypothesized_variance


    p_value = 1 - stats.chi2.cdf(chi_square_statistic, df=sample_size - 1)


    reject_null = p_value < alpha

    return chi_square_statistic, p_value, reject_null


np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
hypothesized_variance = 100

chi_square_statistic, p_value, reject_null = chi_square_variance_test(sample_data, hypothesized_variance)

print(f"Chi-Square Statistic: {chi_square_statistic}")
print(f"P-Value: {p_value}")
print("Reject Null Hypothesis" if reject_null else "Fail to Reject Null Hypothesis")
```

25.  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
import scipy.stats as stats

def z_test_proportions(successes1, n1, successes2, n2, alpha=0.05):
    p1 = successes1 / n1
    p2 = successes2 / n2

    p_pool = (successes1 + successes2) / (n1 + n2)

    std_error = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))


    z_score = (p1 - p2) / std_error


    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))


    reject_null = p_value < alpha

    return z_score, p_value, reject_null

successes1 = 50
n1 = 200
successes2 = 30
n2 = 150

z_score, p_value, reject_null = z_test_proportions(successes1, n1, successes2, n2)

print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")
print("Reject Null Hypothesis" if reject_null else "Fail to Reject Null Hypothesis")
```

26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test(sample1, sample2):

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)


    f_statistic = var1 / var2


    df1 = len(sample1) - 1
    df2 = len(sample2) - 1


    p_value = 1 - stats.f.cdf(f_statistic, df1, df2)
    p_value = 2 * min(p_value, 1 - p_value)

    return f_statistic, p_value, df1, df2


np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=2, size=30)
sample2 = np.random.normal(loc=10, scale=3, size=30)

f_statistic, p_value, df1, df2 = f_test(sample1, sample2)

print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")
print("Reject Null Hypothesis" if p_value < 0.05 else "Fail to Reject Null Hypothesis")


x = np.linspace(0, 5, 1000)
y = stats.f.pdf(x, df1, df2)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})')
plt.axvline(f_statistic, color='red', linestyle='--', label=f'F-Statistic: {f_statistic:.2f}')
plt.title('F-Distribution and Test Statistic')
plt.xlabel('F-Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


np.random.seed(42)
observed_data = np.random.choice([1, 2, 3, 4, 5, 6], size=100, p=[0.1, 0.1, 0.2, 0.2, 0.2, 0.2])

expected_prob = [1/6] * 6
expected_data = np.array(expected_prob) * len(observed_data)


chi2_statistic, p_value = stats.chisquare(f_obs=observed_data, f_exp=expected_data)


print(f"Chi-Square Statistic: {chi2_statistic}")
print(f"P-Value: {p_value}")
if p_value < 0.05:
    print("Reject Null Hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to Reject Null Hypothesis: The observed data fits the expected distribution.")


plt.bar(range(1, 7), np.bincount(observed_data)[1:], align='center', alpha=0.7, label='Observed Data')
plt.plot(range(1, 7), expected_data, 'ro-', label='Expected Data')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.title('Chi-Square Goodness of Fit Test')
plt.legend()
plt.show()
```
